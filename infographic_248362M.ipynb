```python
location_column = 'Location'
death_columns = ['No.of Deaths Male', 'No.of Deaths Female']

# Filter data for Gampaha, Colombo, and Kurunegala
selected_locations = ['Gampaha (Gampaha, Negombo & Kelaniya Divisions)', 
'Colombo (Colombo-South, North, Central, Mt. Lavinia & Nugegoda Divisions)', 
'Kurunegala (Kurunegala, Kuliyapitiya & Nikaweratiya Divisions)']

selected_data = df[df[location_column].isin(selected_locations)]

# Summing deaths for selected locations
total_deaths_selected = selected_data[death_columns].sum().sum()

# Calculate total deaths
total_deaths = df[death_columns].sum().sum()

# Calculate percentage
percentage_deaths = (total_deaths_selected / total_deaths) * 100

print(f"The percentage of deaths that happened in Gampaha, Colombo, and Kurunegala compared to total deaths is: {percentage_deaths:.2f}%")
```

```python
# Selecting relevant columns for vehicle types and deaths
vehicle_columns = ['Motor Car', 'Dual Purpose Vehicle', 'Lorry', 'Cycle', 'Motor Cycle/Moped', 'Three wheeler', 'Articulated Vehicle, prime mover', 'Bus', 'Land Vehicle/Tractor', 'Animal drawn vehicle or rider on animal', 'Other']
death_columns = ['No.of Deaths Male', 'No.of Deaths Female']

# Summing deaths for each vehicle type
total_deaths_per_vehicle = df[vehicle_columns].sum()

# Finding the top 3 vehicle types with the highest death count
top_3_vehicle_types = total_deaths_per_vehicle.nlargest(3)

print("Top 3 vehicle types involved in fatal accidents:")
for idx, vehicle_type in enumerate(top_3_vehicle_types.index, start=1):
    death_count = top_3_vehicle_types[vehicle_type]
    print(f"{idx}. {vehicle_type}: {death_count} deaths")
```
